```python
import requests
from bs4 import BeautifulSoup

def get_survey_numbers(district, mandal, village):
    url = "https://dharani.telangana.gov.in/knowLandStatus"


    response = requests.get(url)


    soup = BeautifulSoup(response.content, "html.parser")


    form_data = {}
    for dropdown in soup.find_all("select"):
        name = dropdown.get("name")
        if name == "ddlDistrict":
            form_data[name] = district
        elif name == "ddlMandal":
            form_data[name] = mandal
        elif name == "ddlVillage":
            form_data[name] = village


    response = requests.post(url, data=form_data)


    soup = BeautifulSoup(response.content, "html.parser")


    survey_numbers = []
    for option in soup.find_all("option"):
        survey_numbers.append(option.text.strip())

    return survey_numbers
```

```python
district = "Adilabad"
mandal = "Bela"
village = "Datalpur"

survey_numbers = get_survey_numbers(district, mandal, village)
print("Survey Numbers:")
for number in survey_numbers:
    print(number)
```
